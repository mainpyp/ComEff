```python
# Normal Plot
import matplotlib.pyplot as plt
import numpy as np
import math

# show all available styles and apply it
print(plt.style.available)
plt.style.use('ggplot')


x = [x for x in range(10000)]
y1 = [math.sqrt(y) for y in range(10000)]
y2 = [-math.sqrt(y) for y in range(10000)]

# formate string = '[marker][line][color]'
#    marker are dots on each datapoint
#    'b'    blue markers with default shape
#    'or'   red circles
#    '-g'   green solid line
#    '--'   dashed line with default color
#    '^k:'  black triangle_up markers connected by a dotted line

#plt.plot(x, y1, color='c', label='positive', linewidth=3., linestyle='-.')
#plt.plot(x,y2, color='red', label='negative', linewidth=3., linestyle=':')
plt.plot(x, y1, label='positive')
plt.plot(x,y2, label='negative')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Bar Plot with comparision
import matplotlib.pyplot as plt
import random as rn
import numpy as np
import math

x = [x for x in range(10)]
y1 = [rn.randint(0, 20) for y in range(10)]
y2 = [rn.randint(0, 20) for y in range(10)]
y3 = [rn.randint(0, 20) for y in range(10)]

indexes = np.arange(len(x))
offset = 0.25

print(indexes)


# formate string = '[marker][line][color]'
#    marker are dots on each datapoint
#    'b'    blue markers with default shape
#    'or'   red circles
#    '-g'   green solid line
#    '--'   dashed line with default color
#    '^k:'  black triangle_up markers connected by a dotted line

# offset +, - 
plt.bar(indexes + offset, y1, width=offset, color='c', label='random bar 1')
plt.bar(indexes, y2, width=offset, color='b', label='random bar 2')
plt.bar(indexes - offset, y3, width=offset, color='y', label='random bar 3')

plt.xticks([val for val in x])
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Csv Plot
import matplotlib.pyplot as plt
import random as rn
import numpy as np
import pandas as pd
import math
from collections import Counter

df = pd.read_csv('example_data.csv', sep=";")
ages = df['Age']
genders = df['Gender']

print(ages)
 
gender_counter = Counter()
for gender in genders:
    gender_counter.update([gender])

gender, observed_amount = map(list, zip(*gender_counter.most_common(4)))

print(gender,observed_amount, sep="\n")

plt.bar(gender, observed_amount, width=0.1)
plt.xticks(gender)
plt.tight_layout()
plt.show()
    



print(df)
        
        
def counter_example():
    count = Counter(['A', 'B', 'A', 'C', 'B', 'C']) 
    print(count)  # Counter({'A': 2, 'B': 2, 'C': 2})
    count.update(['A', 'D', 'A'])
    print(count)  # Counter({'A': 4, 'B': 2, 'C': 2, 'D': 1})
    print(count.most_common(2))  # [('A', 4), ('B', 2)]
    
counter_example()
    
    
    
```

```python
# Pie Plot
import matplotlib.pyplot as plt
import random as rn
import numpy as np
import pandas as pd

percentages = [10, 15, 25, 30, 20]
labels = [10, 15, 25, 30, 20]
explode = [0, 0, 0, 0.1, 0]

plt.pie(percentages, labels=labels, wedgeprops={'edgecolor': 'black'}, 
        autopct='%1.f%%', explode=explode, startangle=120)
plt.tight_layout()
plt.show()
```

```python
# stack plots

from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")


minutes = [1, 2, 3, 4, 5, 6, 7, 8, 9]

# columns always have to add up to 8 when decresing
player1 = [8, 6, 5, 5, 4, 2, 1, 0, 4]
player2 = [0, 1, 2, 2, 2, 4, 4, 5, 3]
player3 = [0, 1, 1, 1, 2, 2, 3, 3, 1]

labels = ['player1', 'player2', 'player3']
colors = ['#6d904f', '#fc4f30', '#008fd5']

plt.stackplot(minutes, player1, player2, player3, labels=labels, colors=colors)

plt.legend(loc=(0.07, 0.05))

plt.title("My Awesome Stack Plot")
plt.tight_layout()
plt.show()
```

```python
# Advanced line plots
import pandas as pd
from matplotlib import pyplot as plt

data = pd.read_csv('example_data2.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

plt.fill_between(ages, py_salaries, dev_salaries,
                 where=(py_salaries > dev_salaries),
                 interpolate=True, alpha=0.25, label='Above Avg')

plt.fill_between(ages, py_salaries, dev_salaries,
                 where=(py_salaries <= dev_salaries),
                 interpolate=True, color='red', alpha=0.25, label='Below Avg')

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

```python
# Sin fill between
from matplotlib import pyplot as plt
import math
import numpy as np
import pandas as pd

x = (np.arange(0.0, 20.0, 0.1))
y_sin = np.array([math.sin(y) for y in x])
y_cos = np.array([math.cos(y) for y in x])
zero_vals = [0 for val in x]
overall_median = 0

plt.plot(x, y_sin, color='#575757', linewidth=2.5,)
plt.plot(x, y_cos, color='#2e1212', linewidth=.5,)
plt.plot(x,  zero_vals, color='black', linewidth=1.5, linestyle='-.')

plt.fill_between(x, y_sin, overall_median, where=(y_sin > overall_median), 
                 alpha=.2, color='#fcd523', interpolate=True)

plt.fill_between(x, y_sin, overall_median, where=(y_sin <= overall_median), 
                 alpha=.2, color='#ff7878', interpolate=True)
plt.show()


# yellow = #fcd523
# grey = #575757
```
